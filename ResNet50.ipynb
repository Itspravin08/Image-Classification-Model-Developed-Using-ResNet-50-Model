Keras is the high-level API of TensorFlow 2: an approachable, highly-productive interface for solving machine learning problems, with a focus on modern deep learning.
```python
from tensorflow.keras.applications.resnet50 import ResNet50      # CNN Trained Model
from tensorflow.keras.applications.resnet50 import decode_predictions # used to Decode the Prediction Result
from keras.applications.resnet import preprocess_input      # Input preprocess
```

##**Import Lib**

```python
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os
from os import listdir
from PIL import Image as PImage
```

```python
# Decler height and width of images
img_width , img_heigth = 224 , 224
```

##**ResNet-50 Model**
ResNet-50 Model. The ResNet-50 model consists of 5 stages each with a convolution and Identity block. Each convolution block has 3 convolution layers and each identity block also has 3 convolution layers. The ResNet-50 has over 23 million trainable parametersesNet-50 Model. The ResNet-50 model consists of 5 stages each with a convolution and Identity block. Each convolution block has 3 convolution layers and each identity block also has 3 convolution layers. The ResNet-50 has over 23 million trainable parameters



```python
# Download ResNet50 Model

Model_PreTrained = ResNet50(weights='imagenet' , include_top = True , input_shape=(img_width,img_heigth,3))
```

#### **insert current path of image to Predict**

```python
img_path = '/content/39.JPG'
img = image.load_img(img_path,target_size=(img_width,img_heigth))
img_data = image.img_to_array(img)
img_data = np.expand_dims(img_data,axis=0)
img_data = preprocess_input(img_data)
```

##**Predict the Result**

```python
CNN_feacture = Model_PreTrained.predict(img_data , verbose=0)
```

###**Decode the result in the list of tuple(Class, discripation ,probability)**

```python
label = decode_predictions(CNN_feacture)
label = label[0][0]
```

```python
plt.imshow(img)

stringprint = "%.1f" % round(label[2]*100,1)
plt.title(label[1]+ " " + str(stringprint)+ "%" , fontsize=20)
plt.axis('off')
plt.show()
```

```python
# Decode the result in the list of tuple(Class, discripation ,probability)
print("Predcation",decode_predictions(CNN_feacture,top=3)[0])
```
